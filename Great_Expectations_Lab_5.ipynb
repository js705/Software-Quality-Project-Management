```python
import pandas as pd
import great_expectations as gx
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",
                 names=["age", "workclass", "fnlwgt", "education", "education-num", "marital-status",
                        "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss",
                        "hours-per-week", "native-country", "income"])
```

```python
df.head()
```

```python
context = gx.get_context()
data_source = context.data_sources.add_pandas("pandas")
data_asset = data_source.add_dataframe_asset(name="pd dataframe asset")
```

```python
batch_definition = data_asset.add_batch_definition_whole_dataframe("batch definition")
batch = batch_definition.get_batch(batch_parameters={"dataframe": df})
```

```python
expectation = gx.expectations.ExpectColumnValuesToBeBetween(
    column="education-num", min_value=0, max_value=20
)
```

```python
validation_result = batch.validate(expectation)
print(validation_result)
```

```python
expectation_education = gx.expectations.ExpectColumnValuesToBeInSet(
    column="education", value_set=["Bachelors", "HS-grad", "Masters", "PhD", "11th", "9th"]
)
validation_result = batch.validate(expectation)
print(validation_result)
```

```python
expectation_no_nulls_income = gx.expectations.ExpectColumnValuesToNotBeNull(
    column="income"
)
validation_result = batch.validate(expectation)
print(validation_result)
```
